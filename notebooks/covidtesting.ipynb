```python
import pandas as pd
import altair as alt
from IPython.display import display
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# Read in daily state data
# df = pd.read_csv('../data/covidstatesdaily.csv')
df = pd.read_csv('https://covidtracking.com/api/v1/states/daily.csv')
df.head()
```

```python
df = df[['date','state','positive','negative','total','positiveIncrease','negativeIncrease','totalTestResultsIncrease']]
```

```python
df.head()
```

```python
df = df.sort_values(by='date')

df['positive7'] = df.groupby('state')['positiveIncrease'].transform(lambda x: x.rolling(7, 7).mean())
df['negative7'] = df.groupby('state')['negativeIncrease'].transform(lambda x: x.rolling(7, 7).mean())
df['total7'] = df.groupby('state')['totalTestResultsIncrease'].transform(lambda x: x.rolling(7, 7).mean())

df = df.sort_values(by='date', ascending=False)
df[df.state=='GA'].head(10)
```

```python
df['positiveTotalRate'] = (df['positive'] / df['total']).fillna(0)
df['positiveDailyRate'] = (df['positiveIncrease'] / df['totalTestResultsIncrease']).fillna(0)
df['positiveRate7'] = (df['positive7'] / df['total7'])
```

```python
df_GA = df[df.state == 'GA'].sort_values(by='date', ascending=False)#.fillna(0)
mindate_over500 = df_GA[df_GA.total >= 500].sort_values(by='date')[['date']].min().values[0]
df_GA = df_GA[df_GA.date >= mindate_over500]
```

```python
def state_rate(state_abbr):
    minp = 0
    maxp = .4
    
    df_state = df[df.state == state_abbr].sort_values(by='date', ascending=False)#.fillna(0)
    mindate_over500 = df_state[df_state.total >= 500].sort_values(by='date')[['date']].min().values[0]
    df_state = df_state[df_state.date >= mindate_over500]

    source = pd.melt(df_state, id_vars=['date'], value_vars=['positiveTotalRate','positiveRate7'])
    source['value'] = source['value'].apply(lambda x: np.nan if (x< minp or x > maxp) else x)

    source['date'] = pd.to_datetime(source['date'], format='%Y%m%d')
    
    display(df_state.head(14))

    lines = alt.Chart(source).mark_line().encode(
        x=alt.X('date:T', title='date'),
        y=alt.Y('value:Q', axis=alt.Axis(format='%')),
        color=alt.Color('variable:O', scale=alt.Scale(scheme='dark2')),
        strokeDash='variable',
        #tooltip=['date', 'value', 'variable']
    ).properties(
        width=800, height=500,
        title='Rate for {}'.format(state_abbr))
    
    dots = alt.Chart(source).mark_circle().encode(
        x=alt.X('date:T', title='date'),
        y=alt.Y('value:Q', axis=alt.Axis(format='%')),
        color=alt.Color('variable:O', scale=alt.Scale(scheme='dark2')),
        tooltip=['date', alt.Tooltip('value', format='.2%'), 'variable']
    ).properties(
        width=800, height=500,
        title='Positive Test Rates for {}'.format(state_abbr))
    return lines + dots
```

```python
state_rate('GA')
```

```python
rate = 'positiveRate7' # positiveTotalRate
minp = 0
maxp = .2

source = df[(df.date >= 20200401) & (df.state.isin(['GA','SC', 'NC','FL']))][['date','state', rate]]
source = pd.melt(source, id_vars=['date','state'], value_vars=[rate])
source['value'] = source['value'].apply(lambda x: np.nan if (x< minp or x > maxp) else x)
source['date'] = pd.to_datetime(source['date'], format='%Y%m%d')

alt.Chart(source).mark_line().encode(
    x=alt.X('date:T', title='date'),
    y=alt.Y('value:Q', axis=alt.Axis(format='%')),
    color=alt.Color('state:O', scale=alt.Scale(scheme='dark2')),
    strokeDash='state',
    tooltip=['date', alt.Tooltip('value', format='.2%'), 'state']
).properties(
    width=800, height=500,
    title='Rate for {}'.format('7 Day Rolling Average Positive Test Rate'))
```
